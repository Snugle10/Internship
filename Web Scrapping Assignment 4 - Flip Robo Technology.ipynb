### Q1 Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
A)Rank
B) Name
C) Artist
D) Upload date
E) Views 

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]
```

```python
Rank_tags = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for i in Rank_tags:
    number=i.text
    Rank.append(number)
```

```python
Name_tags = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
for i in Name_tags:
    tag_name=i.text
    Name.append(tag_name)
```

```python
artist_tags = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')  
for i in artist_tags:
    tag_artist=i.text
    Artist.append(tag_artist)
```

```python
views_tags = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')  
for i in views_tags:
    watch=i.text
    Views.append(watch)
```

```python
upload_tags = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')  
for i in upload_tags:
    upload=i.text
    Upload_date.append(upload)
```

```python
print(len(Rank),len(Name),len(Artist),len(Views),len(Upload_date))
```

```python
df = pd.DataFrame({'Rank':Rank[:30],'Name':Name[:30],'Artist':Artist[:30],'Upload_date':Upload_date[:30],'Views':Views[:30]})
df
```

```python
driver.close()
```
